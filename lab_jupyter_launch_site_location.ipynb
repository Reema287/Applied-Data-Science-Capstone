```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon

# نقطة مركز الخريطة (NASA Johnson Space Center)
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# إنشاء خريطة جديدة
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# إنشاء MarkerCluster لتجميع العلامات
marker_cluster = MarkerCluster()

# إنشاء عمود جديد marker_color حسب حالة الإطلاق
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# إضافة العلامات لكل عملية إطلاق
for index, row in spacex_df.iterrows():
    # موقع الإطلاق (خط العرض والطول)
    location = [row['Lat'], row['Long']]
    
    # نص يظهر عند الضغط على العلامة
    popup_text = ("Launch Site: {}<br>"
                  "Payload Mass: {} kg<br>"
                  "Launch Outcome: {}").format(
                      row['Launch Site'],
                      row['PayloadMass'] if 'PayloadMass' in row else 'N/A',
                      'Success' if row['class'] == 1 else 'Failure'
                  )
    
    # إنشاء العلامة مع اللون المناسب
    marker = folium.Marker(
        location=location,
        popup=popup_text,
        icon=folium.Icon(color=row['marker_color'])
    )
    
    # إضافة العلامة لمجموعة التجميع
    marker_cluster.add_child(marker)

# إضافة MarkerCluster للخريطة
site_map.add_child(marker_cluster)

# عرض الخريطة
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

# دالة لحساب المسافة بين نقطتين جغرافياً باستخدام Haversine formula
def haversine(coord1, coord2):
    # إحداثيات النقطة الأولى
    lat1, lon1 = coord1
    # إحداثيات النقطة الثانية
    lat2, lon2 = coord2
    
    # تحويل الدرجات إلى راديان
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # فروق الإحداثيات
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # تطبيق صيغة Haversine
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # نصف قطر الأرض بالكيلومتر
    R = 6371
    
    # المسافة النهائية
    distance = R * c
    return distance

# استخراج بيانات المواقع (اسم الموقع والإحداثيات)
launch_sites = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# دالة لحساب المسافات من موقع معين إلى باقي المواقع
def calculate_distances(site_name, sites_df):
    distances = {}
    site_coord = sites_df[sites_df['Launch Site'] == site_name][['Lat', 'Long']].values[0]
    
    for index, row in sites_df.iterrows():
        if row['Launch Site'] != site_name:
            other_coord = (row['Lat'], row['Long'])
            dist_km = haversine(site_coord, other_coord)
            distances[row['Launch Site']] = dist_km
    return distances

# حساب المسافات لكل موقع إطلاق
all_distances = {}

for site in launch_sites['Launch Site']:
    all_distances[site] = calculate_distances(site, launch_sites)

# عرض النتائج
for site, distances in all_distances.items():
    print(f"Distances from {site}:")
    for other_site, dist in distances.items():
        print(f"  To {other_site}: {dist:.2f} km")
    print()

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import math

def haversine_distance(coord1, coord2):
    # إحداثيات النقاط (lat, lon) بالدرجات
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    
    # تحويل الدرجات إلى راديان
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # حساب الفرق بين الإحداثيات
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # تطبيق معادلة هافرسين
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # نصف قطر الأرض بالكيلومتر
    R = 6371.0
    
    # حساب المسافة
    distance = R * c
    return distance
    from folium.features import DivIcon

def add_marker_and_line(map_object, launch_coord, target_coord, label):
    # حساب المسافة باستخدام دالة هافرسين
    distance = haversine_distance(launch_coord, target_coord)

    # إضافة marker مع عرض المسافة
    distance_marker = folium.Marker(
        location=target_coord,
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 14px; color:green;"><b>{label}: {distance:.2f} KM</b></div>'
        )
    )
    distance_marker.add_to(map_object)

    # رسم خط بين الموقعين
    line = folium.PolyLine(locations=[launch_coord, target_coord], weight=3, color='blue')
    line.add_to(map_object)

# مثال: إحداثيات موقع الإطلاق (Launch Site)
launch_site_coord = [28.5623, -80.5774]  # غيّرها حسب موقعك

# مثال: إحداثيات نقطة على الخريطة (مثلاً أقرب مدينة)
closest_city_coord = [28.5791, -80.6457]

# إنشاء الخريطة
site_map = folium.Map(location=launch_site_coord, zoom_start=10)

# إضافة marker وخط
add_marker_and_line(site_map, launch_site_coord, closest_city_coord, 'Closest City')

# عرض الخريطة
site_map

```
